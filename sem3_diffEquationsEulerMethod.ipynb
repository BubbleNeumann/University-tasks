```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
import numpy as np

fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x, y)

ax.plot(x, np.sin(x), color=(1.0,0.2,0.3), linestyle='--')

ax.plot(x, x+3, color='blue', linestyle=':')

ax.plot(x, x, color='#FFDD44', linestyle='-.')
```

```python
fig, ax = plt.subplots()
x = np.array([-3, -2, -1, 0, 1, 2, 3])

ax.set_yticks(np.arange(0, 18, 1))

ax.plot(x, x**2, marker='o')

ax.plot(x, 2 * x**2, marker='o', alpha=0.2)

ax.plot(0.5 * x, x**2, marker='o', linewidth=2.5)
```

```python
fig, ax = plt.subplots()
x = np.array([-3, -2, -1, -0.5, -1, 0, 0.5, 1, 2, 3])

ax.plot(x, x**2, color='red', marker='o', label='y = x^2')

ax.set_title('the title', fontsize=16)

ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax.legend(loc='lower right')

ax.set_xlim(0, 4)
ax.set_ylim(-3, 10)

ax.grid(color='blue', linewidth=0.5, alpha= 0.2)
# to remove grid use ax.grid()
```
